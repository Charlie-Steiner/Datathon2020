```python
def getIndustrySales(source,num,name):
    source=source.get(["GEO.id2","NAICS.id","RCPSZFE.id","ESTAB"])
    source=source[source.applymap(isnumber)]
    source=source.astype('int32')
    source=source[source['NAICS.id']==num].rename(columns={"GEO.id2":"zipcode"})
    source=source.pivot_table(index='zipcode',columns='RCPSZFE.id',values='ESTAB',fill_value=0)
    source[name]=source.get([114])*SalesEstimate[114]
    source[name]=source.get(name)+source.get(123)*SalesEstimate[123]
    source[name]=source.get(name)+source.get(125)*SalesEstimate[125]
    source[name]=source.get(name)+source.get(131)*SalesEstimate[131]
    source[name]=source.get(name)+source.get(132)*SalesEstimate[132]
    source=source.drop(columns=[1,2,114,123,125,131,132,998])
    print(source.head(5))
    return source
```

```python
def ProcessColumns(presource):
    presource=presource[presource['NAICS'].map(lambda x: len(x)) == 6] 
    presource=presource.astype({"zipcode":'int32','NAICS':"int32","RCP":'int32',"Number":'int32'})
    source=presource.pivot_table(index='zipcode',columns=['NAICS','RCP'],values='Number',fill_value=0)
    for (a,b) in source.columns:
        if b==114:
            source[(a,1)]=source.get((a,b))*SalesEstimate[114]
        elif b==123:
            source[(a,1)]+=source.get((a,b))*SalesEstimate[123]
        elif b==125:
            source[(a,1)]+=source.get((a,b))*SalesEstimate[125]
        elif b==131:
            source[(a,1)]+=source.get((a,b))*SalesEstimate[131]
        elif b==132:
            source[(a,1)]+=source.get((a,b))*SalesEstimate[132]
    target = pd.DataFrame()
    NAICS = [a for (a,b) in source.columns]
    for n in NAICS:
        target[n]=source.get((n,1))
    return target
```

```python
df2t=ProcessColumns(df2)
```

```python
df3t=ProcessColumns(df3)
df4at=ProcessColumns(df4a)
df4bt=ProcessColumns(df4b)
df5t=ProcessColumns(df5)
```
