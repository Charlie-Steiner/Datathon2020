## PCA Projection to 2D

```python
pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(x2)
```

```python
def getFirst(t):
    return t[0]
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = np.arange(2))
top2=[()]*108
for a in np.arange(108):
    top2[a]=(principalDf[0][a], principalDf[1][a], df.columns[a])

top2.sort(key=getFirst)

print(top2)
```

```python
ID=0
target=pd.DataFrame(False, index=np.arange(len(df[features[ID]])), columns=[features[ID]+"big"])
for a in np.arange(len(df[features[ID]])):
    target[features[ID]+"big"][a]=(df2[features[ID]][a]>0)
target.head(5)
```

```python
finalDf = pd.concat([principalDf, target],axis=1)
finalDf.head(5)
```

## Visualize 2D Projection

Use a PCA projection to 2d to visualize the entire data set. You should plot different classes using different colors or shapes. Do the classes seem well-separated from each other? 

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = [True, False]
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf[features[ID]+"big"] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
